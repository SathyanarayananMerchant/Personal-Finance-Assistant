```python
def handle_outliers(data, column, method="remove"):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    if method == "remove":
        
        return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    elif method == "mean":
        
        mean_value = data[column].mean()
        data[column] = np.where((data[column] < lower_bound) | (data[column] > upper_bound), mean_value, data[column])
        return data
    elif method == "median":
        
        median_value = data[column].median()
        data[column] = np.where((data[column] < lower_bound) | (data[column] > upper_bound), median_value, data[column])
        return data
    else:
        raise ValueError("Method should be 'remove', 'mean', or 'median'.")
```

```python
handling_method = "median"
```

```python
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    df = handle_outliers(df, column, method=handling_method)
```
